```
import pandas as pd
import pyspark
```

```
data = pd.read_csv('https://firms.modaps.eosdis.nasa.gov/data/active_fire/modis-c6.1/csv/MODIS_C6_1_Global_24h.csv')
data
```

```
filtered_data = data[(data['latitude'] > 40) & (data['latitude'] < 55) & (data['longitude'] > 10) & (data['longitude'] < 25)]
filtered_data
```
